```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('organizationDF.tsv', sep='\t')
data = data.drop('Unnamed: 0', 1)

# Parse only funded EdTech startups
data = data[data.funding != 0]
data.head(1)
```

#### Funding 
To my dismay, it appears that a key statistic, the quantity in USD that a company received in funding,
has many holes. Almost 1000 companies have a count of 0 for funding.

Although I would rather use a tag such as 'education', which would undoubtedly provide better results,
with tens of thousands of education companies I wouldn't be able to process the data via the API rate limits.
'EdTech' still provides a focused view of hundreds of companies that received funding, and ultimately
gives a more intimate perspective of this market.

Despite the fact that many companies don't have funding information, I will work around this problem
by focusing my analysis upon the companies that do have this information. And in the same method of focusing
my approach to the companies with funding information, I will compare this subset of companies on 
other measures such as date founded, number of employees, number of acquisitions, and ipo.

```python
# Analyze funding data
funding = data['funding'].value_counts()
funding = funding.sort_index(ascending=False)
plt.show()

to500 = 0
to1mm = 0
to5mm = 0
to10mm = 0
to100mm = 0
rest = 0

# Group funding into categories
for index in funding.index:
    if index <= 500000:
        to500 += funding[index]
    if index > 500000 and index <= 1000000:
        to1mm += funding[index]
    if index > 1000000 and index <= 5000000:
        to5mm += funding[index]
    if index > 5000000 and index <= 10000000:
        to10mm += funding[index]
    if index > 10000000 and index <= 100000000:
        to100mm += funding[index]
    if index > 100000000:
        rest += funding[index]
    
# Plot funding Bargraph
fig = plt.figure(figsize=(15,10))
bars = plt.bar([2,3,4,5,6,7], [to500, to1mm, to5mm, to10mm, to100mm, rest])
plt.xticks([2,3,4,5,6,7],['0-50k', '50k-1mm', '1mm-5mm', '5mm-10mm', '10mm-100mm', '100mm+'])
fig.suptitle('Total Funding in USD (x) by Number of Companies (y)', fontsize=18)
plt.tick_params(axis='both', which='major', labelsize=15)

colors = ['r', 'g', 'y', 'k', 'm', 'b']
for i in range(0, len(bars)):
    bars[i].set_color(colors[i])
```

```python
# Analyze funding rounds data
fig = plt.figure(figsize=(10,10))
funding_rounds = data['funding_rounds'].value_counts()
funding_rounds.plot(kind='barh')
fig.suptitle('Number of Companies (x) by Number of Funding Rounds (y)', fontsize=18)
plt.tick_params(axis='both', which='major', labelsize=15)

# Number of Funding Rounds
```

#### Location
Our dataset is international, and as such, we get a snapshot of the internation EdTech startup scene.
Surprisingly, New York City represents the largest number of funded EdTech startups, with San Francisco in 
a close second. But the Sillicon Valley area encompasses not only San Francisco, but San Mateo and Palo Alto.
Combining these 3 areas together we can see that, as expected, Sillicon Valley hosts the largest number of
funded EdTech startups in the world. 

After Sillicon Valley and New York, Beijing and London host an impressive number of funded EdTech companies
as well. Boston, MA comes in next to round out the top 5 funded EdTech startup hubs of the world.

```python
# Analyze Location data
location = data.location.value_counts()
location.sort_index()

california = 0
nyc = 0
boston = 0
beijing = 0
london = 0
restUS = 0
restInternational = 0

# Group data by Location (changed to state - variable names don't reflect this)
for index in location.index:
    if 'New York' in index:
        nyc += location[index]
        continue
    if 'California' in index:
        california += location[index]
        continue
    if 'China' in index:
        beijing += location[index]
        continue
    if 'England' in index:
        london += location[index]
        continue
    if 'Massachusetts' in index:
        boston += location[index]
        continue
    if 'United States' in index:
        restUS += location[index]
    else: 
        restInternational += location[index]

# Plot bar graph
fig = plt.figure(figsize=(10,10))
bars = plt.bar([1,2,3,4,5,6,7], [california, nyc, boston, beijing, london, restUS, restInternational])
plt.xticks([1,2,3,4,5,6,7],['CA', 'NY', 'MA', 'China', 'England', 'US', 'International'])
fig.suptitle('Location (x) by Number of Companies (y)', fontsize=18)
plt.tick_params(axis='both', which='major', labelsize=15)

colors = ['r', 'g', 'y', 'k', 'm', 'b', 'c']
for i in range(0, len(bars)):
    bars[i].set_color(colors[i])

plt.show()
```

#### Number of Employees
To keep the type of company that is representative of a funded EdTech startup in perspective,
it is useful to look at the number of employees. A large quantity of these companies
don't report the number of employees - and we can assume these companies employee only a handful of people. 

Most of these companies employ 10 or less people, although there are some relative outliers, with one company employing 1000
people and a few other employing in the 100's.

```python
# Analyze Number of Employee data
num_employees = data.num_employees.value_counts()
num_employees = num_employees.sort_index()

to10 = 0
to50 = 0
to100 = 0
to200 = 0
to1000 = 0
for index in num_employees.index:
    if index < 11:
        to10 += num_employees[index]
    if index > 10 and index < 51:
        to50 += num_employees[index]
    if index > 50 and index < 101:
        to100 += num_employees[index]
    if index > 100 and index < 201:
        to200 += num_employees[index]
    if index > 200:
        to1000 += num_employees[index]
        
fig = plt.figure(figsize=(10,10))
bars = plt.bar([1,2,3,4,5], [to10, to50, to100, to200, to1000])
plt.xticks([1,2,3,4,5],['0-10','11-50','51-100','101-200','201-1000'])
fig.suptitle('Number of Employees (x) by Number of Companies (y)', fontsize=20)
plt.tick_params(axis='both', which='major', labelsize=15)

colors = ['r', 'g', 'y', 'k', 'm', 'b', 'c']
for i in range(0, len(bars)):
    bars[i].set_color(colors[i])
    
# num_employees.plot(kind='barh', figsize=(10,10))
```

```python
# Analyze Funding vs Number of Employees
xcol = []
ycol = []
ycol2 = []
for idx, row in data.iterrows():
    xcol.append(row.funding)
    ycol.append(row.num_employees)
    ycol2.append(row.funding_rounds)

clusters_x = []
clusters_y = []
for i in range(0, 4):
    clusters_x.append([])
    clusters_y.append([])

for i in range(0, len(xcol)):
    if ycol[i] < 50:
        clusters_x[0].append(xcol[i])
        clusters_y[0].append(ycol[i])
        continue
    if ycol[i] < 100:
        clusters_x[1].append(xcol[i])
        clusters_y[1].append(ycol[i])
        continue
    if ycol[i] < 200:
        clusters_x[2].append(xcol[i])
        clusters_y[2].append(ycol[i])
        continue
    if ycol[i] >= 200:
        clusters_x[3].append(xcol[i])
        clusters_y[3].append(ycol[i])
        continue

        
# Funding by Number of Employees
fig = plt.figure(figsize=(20,10))

colors = ['0.99', '0.01', '0.75', '0.3']
for i in range(0, 4):
    plt.scatter(clusters_x[i], clusters_y[i], c=colors[i], s=200)
    
fig.suptitle('Funding (x) by Number of Employees (y)', fontsize=20)
plt.tick_params(axis='both', which='major', labelsize=15)
```

```python

cluster1_x = []
cluster1_y = []
cluster2_x = []
cluster2_y = []
cluster3_x = []
cluster3_y = []
cluster4_x = []
cluster4_y = []
for i in range(0, len(xcol)):
    if xcol[i] > 50000000:
        cluster4_y.append(ycol2[i])
        cluster4_x.append(xcol[i])
        continue
    if ycol2[i] > 4:
        cluster1_y.append(ycol2[i])
        cluster1_x.append(xcol[i])
        continue
    if ycol2[i] < 3:
        cluster2_y.append(ycol2[i])
        cluster2_x.append(xcol[i])
    if ycol2[i] >= 3 and ycol2[i] < 5:
        cluster3_y.append(ycol2[i])
        cluster3_x.append(xcol[i])

# Funding by Number of Funding Rounds
fig = plt.figure(figsize=(20,10))
plt.scatter(cluster1_x, cluster1_y,c='0.02', s=200)
plt.scatter(cluster2_x, cluster2_y,c='0.75', s=200)
plt.scatter(cluster3_x, cluster3_y,c='0.99', s=200)
plt.scatter(cluster4_x, cluster4_y,c='0.4', s=200)
fig.suptitle('Funding (x) by Number of Funding Rounds (y)', fontsize=20)
plt.tick_params(axis='both', which='major', labelsize=15)
```

### Funding Graphs
We can draw some conclusions on the nature of the funded EdTech startups we are analyzing from the above graphs.
We are trying to learn what identifies a succesful EdTech startup, and these comparisons can help us do so.

**Funding vs Number of Employees**

Although there is a hole in the data for number of employees (at least for a significant number of the companies
being observed), we can see the surprising result that number of employees doesn't necessarily correlate with total funding.
At the same time we can see a "sweet spot" for potentially succesful companies at around 150-400 employees. 
This limited group raises significant funding while having hundreds of employees - from these measures it seems likely
the company is having sustained growth. It is important to note that of the companies with hundreds of employees, some
may refuse funding because they have a reliable revenue stream.

The majority of the EdTech startups we are observing can be found on the x-axis between 0.0-0.5, having received less than 1MM in funding.

**Funding vs Number of Funding Rounds**

Looking at the above graph, it is clear that the number of funding rounds a company is correlated to a company's
overall funding (and likely, its success as well). Again, the majority of EdTech startups are clumped together with less than
1MM funding, ranging in 1-3 rounds of funding. And again, the companies we are most interested in, those potentially very succesful, are found at around 3-6 funding rounds, with considerable overall funding.


```python
from __future__ import division

success_funding = data[data.num_employees > 0]
success_funding['funding_ratio'] = pd.Series(success_funding.funding / success_funding.num_employees)

xcol = []
xcol2 = []
ycol = []
ycol2 = []
success = []
for idx, row in success_funding.iterrows():
    # single outlier ($10.6mm per employee)
    if row.funding_ratio == 10600000.0:
        success.append(row.name)
        continue
    # Cluster most succesful companies
    years_operating = 2015 - int(row.founded[:4])
    if years_operating >= 10 and years_operating != 2015 or row.funding_ratio > 500000:
        if int(row.founded[:4]) == 0:
            ycol2.append(1)
        else:
            ycol2.append(years_operating)
        xcol2.append(row.funding_ratio)
        success.append(row.name)
        continue
    # Cluster less succesful companies
    if int(row.founded[:4]) == 0:
        ycol.append(1)
    else:
        ycol.append(years_operating)
    xcol.append(row.funding_ratio)

    
# funding / # employees (x) by years operating (y)
fig = plt.figure(figsize=(10,10))
plt.scatter(xcol, ycol, c='.01', s=200)
plt.scatter(xcol2, ycol2, c='.5', s=200)

fig.suptitle('Funding Ratio (x) by Number of Years Operating (y)', fontsize=20)
plt.tick_params(axis='both', which='major', labelsize=15)
```

```python
import statsmodels.api as sm

# Linear Regression on Successful companies (grey cluster above)
model = sm.OLS(ycol2, xcol2)
results = model.fit()
print results.summary()
print "Confidence Intervals:", results.conf_int()
print "Parameters:", results.params
```

```python
companies_df = pd.DataFrame(columns=
                            ['name','location','founded','funding', 'funding_rounds', 'ipo', 'num_employees', 'funding_ratio', 'years_operating'])

i = 0
for idx, row in success_funding.iterrows():
    if idx in success:
        row['years_operating'] = 2015 - int(row.founded[:4])
        row['location'] = row.location.split(',',1)[1]
        companies_df.loc[i] = row
        i += 1 

# Funding ratio extremes
max_ratio = companies_df[companies_df.funding_ratio == np.amax(np.array(companies_df.funding_ratio))]
min_ratio = companies_df[companies_df.funding_ratio == np.amin(np.array(companies_df.funding_ratio))]
# Longevity extremes
max_years = companies_df[companies_df.years_operating == np.amax(np.array(companies_df.years_operating))]
min_years = companies_df[companies_df.years_operating == np.amin(np.array(companies_df.years_operating))]

print 'Funding Ratio Max: '
print '-' * 20
print max_ratio
print '=' * 60
print 'Funding Ratio Min: '
print '-' * 20
print min_ratio
print '=' * 60
print 'Longevity Max: '
print '-' * 20
print min_years
print '=' * 60
print 'Longevity Min: '
print '-' * 20
print max_years
print '=' * 60


funding_ratios = np.array(companies_df.funding_ratio)
print 'Average Funding Ratio (total funding USD / # employees): ', int(np.mean(funding_ratios))
print '=' * 60

num_employees = np.array(companies_df.num_employees)
print 'Average Number of Employees: ', int(np.mean(num_employees))
print '=' * 60

funding_rounds = np.array(companies_df.funding_rounds)
print 'Average Number of Funding Rounds: '+ "{0:.2f}".format(np.mean(funding_rounds))
print '=' * 60

years_operating = np.array(companies_df.years_operating)
print 'Average Number of Years Operating: '+ "{0:.2f}".format(np.mean(years_operating))
print '=' * 60

print 'Top Locations:'
pd.set_option('display.max_rows', 5)
print companies_df.location.value_counts()[:5]
```
