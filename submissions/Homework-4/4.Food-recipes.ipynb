## Trial 1 - Agglomerative Clustering (with Dendrogram)
------------------------------------------------------

The goal of my initial clustering was to try and cluster recipes (lines from the .txt file)
based opon shared ingredients. In doing so, the most common ingredients regardless of
each recipe's origin could be discovered.

First, I took each recipe as an individual element.
Using the aggregated vocabulary, which was a list of all potential ingredients,
I did a tf-idf vectorizing of the list of recipes. In doing so, I was able to
calculate the frequency of ingredients in each recipe.

After this, I calculated the euclidean distances between recipes using pdist().
This data was then formatted properly, allowing for an agglomerative clustering using linkage().
With said clustering, I was able to produce a dendrogram. But unfortunately, the resulting
figure does not provide many valuable insights to the data. 

The hierarchy of the dendrogram serves to show the clusterings of similar recipes,
regardless of origin. Recipes that share ingredients are placed in the same cluster.

It is of note that there are a number of recipes sharing a particular ingredient (column around value 1.4),
but from our clustering we are unable to see the textual name of this ingredient.
The dendrogram is practically flat other than this, hovering slightly above the value of 1.
With little seperation between groupings of recipes, the dendrogram doesn't provide many other insights.

Although the data was effectively grouped on similarity, the resulting visualization was unable
to convey a substantial message. With help from proper categorical labeling or ingredient
listing, the dendrogram would be far more useful. These are the primary points of improvement
I sought out while taking a similar, but adapted approach to my second clustering using K-means.


--------------------------------------------------------


```python
import pandas as pd
import numpy as np
import sklearn.cluster as cluster
import sklearn.feature_extraction.text as tf_idf
import seaborn as sb
import scipy.cluster.hierarchy as hr
from scipy.spatial.distance import pdist
import matplotlib.pyplot as plt
%matplotlib inline
import sys
sys.setrecursionlimit(10000)


# build set of all ingredients
recipes = []
words = {}
for line in open("./epic_recipes.txt", 'r'):
    recipes.append(line)
    line = line.split('\t')[1:]
    for word in line:
        if '\n' in word:
            word = word[:-1]
        if not word in words:
            words[word] = True
vocab = []
for key in words:
    vocab.append(key)

# vectorize array of recipes on the basis of all possible ingredients 
# (excludes category e.g. Southwestern)
vectorizer = tf_idf.TfidfVectorizer("./epic_recipes.txt", vocabulary=vocab)
matrix_data = vectorizer.fit_transform(recipes).toarray()

# calculate euclidean distances of vectorized text file
dist_matrix = pdist(matrix_data)
```

```python

plt.figure(figsize=(100,100))

# do an agglomerative clustering on the euclidean distance matrix
linked = hr.linkage(dist_matrix)
```

```python
# display a dendrogram of the agglomerative clustering
dendro = hr.dendrogram(linked,
                       truncate_mode='lastp',
                       show_leaf_counts=True,
                       color_threshold=1)


plt.show()
```

I then displayed a bar graph for each category's top ten ingredients.

While not as powerful a tool as the visualization a clustering can provide,

But the best insights to be gathered from these graphs come from comparing

```python
from sklearn.cluster import KMeans

ingredientsByCategory = {}
for line in open("./epic_recipes.txt", 'r'):
    ingredientCounter = {}
    line = line.split('\t')
    category = line[0]
    # create a list of recipes for each category
    if not category in ingredientsByCategory:
        ingredientsByCategory[category] = {}
    for ingredient in line[1:]:
        if '\n' in ingredient:
            ingredient = ingredient[:-1]
        if ingredient not in ingredientsByCategory[category]:
            ingredientsByCategory[category][ingredient] = 0
        ingredientsByCategory[category][ingredient] += 1
    # add recipe to category



mostIngredients = 0
categories = {}
for category in ingredientsByCategory:
    if len(ingredientsByCategory[category]) > mostIngredients:
        mostIngredients = len(ingredientsByCategory[category])
    fig = plt.figure()
    ax = fig.add_subplot(1,1,1)
    ax.set_title(category)
    topTen = []
    i = 0
    for ingredient in sorted(ingredientsByCategory[category], key=ingredientsByCategory[category].get, reverse=True):
        # gather top 10 most frequently used ingredients
        # for each category of food
        if i < 10:
            topTen.append((ingredient, ingredientsByCategory[category][ingredient]))
            i += 1
        categories[category] = topTen
        
    counts = []
    ingredients = []
    for ingredient, count in topTen:
        ingredients.append(ingredient)
        counts.append(count)
    ax.barh(np.arange(10), counts, 0.5)
    ax.set_yticks(range(0,10))
    xTicks = ax.set_yticklabels(ingredients)
    
    
plt.show()
    

        

```

## Trial 3 - Hierarchical Clustering by Category
------------------------------------------------

For my third and final trial, I utilized the data that had been analyzed
in my second trial. With the counts of all ingredients for each category,
I sought out to examine the total number of ingredients found in all
recipes for each category, as well as how well each ingredient was represented.

The information on intersecting ingredients (e.g. Thai and Mexican using cayenne frequently)
was already gathered in Trial two, so I decided to perform a hierarchical clustering on
the data. This time around, the clustering performed better.

From the resulting dendrogram, we can see the categories of cuisine that represent
the highest number of ingredients from their recipes. Clearly, American cuisine
has the largest ingredient count. Italian, Asian, French and Mexican all follow 
closely, but there is a significant dropoff and leveling out after that.
