## Step 1. Simple clustering

The first thing that we will try is to cluster the geolocation data of the calls. The geolocation of a record consists of a pair `(latitude, longitude)`, which represents the position of the record on the map. Since the dataset that we have spans only a small geographic location, i.e. Boston, we will use the Euclidean distance when measuring how close two points are. Techinically, we should pick a [projection](https://github.com/mbostock/d3/wiki/Geo-Projections) in order to transform the data into 2d, and measure  the Euclidean distance in the projected space.

Since we are dealing with numeric data, we might consider centering and normalizing them. In order to do that, check the [StandardScaler](http://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html#sklearn.preprocessing.StandardScaler) from `scikit-learn`.

For exampe, the result of a clustering with 10 clusters is the following:


![A clustering of the locations of the calls for k=10](clusters10.png)

Filter the dataset to get the latitude and longitude, transform it as required and run the k-means clustering algorithm. Argue about the number of clusters you used. **(pts 50)**

```python
from sklearn.preprocessing import StandardScaler
import sklearn.metrics as metrics
import random
import numpy as np
# avoid int division
from __future__ import division
import matplotlib.pyplot as plt

std = StandardScaler()

# print boston.Geocoded_Location[0]

lat_long = boston.Geocoded_Location

# Convert string lat long data to (float, float)
validData = []
for x in lat_long:
    x = x.replace('(', '').replace(')', '')
    pairArray = x.split(', ')
    pair = (float(pairArray[0]), float(pairArray[1]))
    validData.append(pair)

validData = std.fit_transform(validData)
# print validData


def untilConvergence(k, data, centroids):
    """assign points to clusters, update centroids"""
    clusters = []
    for i in range(0, k):
        clusters.append([])
    
    # find closest centroid
    for point in data:
        euclideanDists = metrics.euclidean_distances(point, centroids)
        minDist = euclideanDists[0][0]
        chosenCentroid = 0
        index = 0
        for distPair in euclideanDists[0]:
            if (np.sum(distPair) < np.sum(minDist)):
                minDist = distPair
                chosenCentroid = index
            index += 1
        # add point to closest cluster (from centroid)
        clusters[chosenCentroid].append(point)

    # update centroids to mean point of respective cluster
    print 'Size of clusters: '
    for i in range(0, k):
        clusterSum = [0, 0]
        print i, ' ', len(clusters[i])
        for point in clusters[i]:
            clusterSum[0] += point[0]
            clusterSum[1] += point[1]
        clusterMean = (clusterSum[0] / len(clusters[i]), clusterSum[1] / len(clusters[i]))
        centroids[i] = np.asarray(clusterMean)
    
    return (clusters, centroids)
    
    
def testConvergence(centroids, oldCentroids):
    """test if any centroids were updated previous iteration"""
    convergenceCount = 0
    
    for i in range(0, len(centroids)):
        if (centroids[i] == oldCentroids[i]).all():
            convergenceCount += 1
    
    if (convergenceCount == len(centroids)):
        return True
    else:
        return False
    
    
def kMeans (k, data):
    """run algorithm until convergence for best clusters"""
    clusters = []
    centroids = []
    
    # initialize clusters
    for i in range(0, k):
        clusters.append([])
        # assign random centroids
        centroids.append(data[random.randint(0, len(data))])
    converged = False
    oldCentroids = None
    while (not converged):
        (clusters, centroids) = untilConvergence(k, data, centroids)
        print 'Current: ',centroids
        print 'Old: ', oldCentroids
        if (oldCentroids):
            converged = testConvergence(centroids, oldCentroids)
        oldCentroids = centroids[:]
    
    return clusters
        

clusteredData = kMeans(7, validData)
print 'Converged'


# Generate a scatter plot for clustered Data
f, axs = plt.subplots(1,1,figsize=(10,10))

xCol = []
yCol = []
colors = []
#     colors = ['0.01', '.2', '.3', '0.5', '.7', '.85', '0.9']

for i in range(0, len(clusteredData)):
    colors.append(str(random.random()))
    for point in clusteredData[i]:
        xCol.append(point[0])
        yCol.append(point[1])
    plt.scatter(xCol, yCol, c=colors[i])
    xCol = []
    yCol = []

    



# I choose to use 7 clusters because for K > 3, the error increases greatly.
# With K = 7, I attempt to achieve enough granularity for the graph to 
# provide utility and increase the viewer's understanding of the data,
# while keeping K as low as possible so as to reduce the percent error.
# Although K = 3 provides signficant improvement of accuracy, the results
# yielded by the resulting graph do little to improve the user's understanding
# of the dataset.

        
```

## Step 2. Using multiple features

The next step in our analysis is to use more features than just the geolocation of a record (whose clustering only gives us information about the neighborhoods, which should not be a huge revelation!) Filter the dataset to get more features, transform them as required and compute another clustering. You should argue about the quality of your clustering and use domain knowledge (i.e. the fact that you are locals) to explore the clusters you produced and present them to the reader. **(pts 30)**

```python
openDate = boston.OPEN_DT
closedDate = boston.CLOSED_DT
closedDate = closedDate.fillna(value='2015-03-01 00:00:01')

# Convert date data to viable format
dateData = []
for i in range(0, len(openDate)):
    tempOpen = openDate[i].replace('-', '').replace(' ', '').replace(':', '')
    tempClose = closedDate[i].replace('-', '').replace(' ', '').replace(':', '')
    pair = (float(tempOpen), float(tempClose))
    dateData.append(pair)

dateData = std.fit_transform(dateData)

clusteredDates = kMeans(3, dateData)

# Generate a scatter plot for clustered Data
f, axs = plt.subplots(1,1,figsize=(20,20))

xCol = []
yCol = []
colors = ['0.85', '0.7', '0.25']

for i in range(0, len(clusteredDates)):
    for point in clusteredDates[i]:
        xCol.append(point[0])
        yCol.append(point[1])
    plt.scatter(xCol, yCol, c=colors[i])
    xCol = []
    yCol = []


# At first glance, this scatter plot may seem to convey little information.
# With only K = 3, the viewer isn't struck with any immediate information.
# But with the context of the data set in mind, the graph can take new meaning.

# Given that the open dates of cases run along the x-axis, and closed on the y-axis,
# we can begin to discover some information. Cases still open were given a filler
# close date (marked as today), as indicated by the lines at y = 1.5.
# But the best insight to be found is that of case frequency.

# Frequency can be determined, with x-axis and y-axis in mind, by the density
# of cases being opened and closed at any value of X. For example, at X = ~ -0.8,
# we can see there was a signficiant number of cases, and as a result, the closing
# dates of many were delayed. The density of these clusters points to this activity,
# and with the context of the data in mind, it can be determined that as this cluster's
# X position signifies that it is a month (namely January), we can see from this data that 
# a large snow storm occured in the final days of January, due to the number of Opened/Closed cases.
```

## Step 3. Detecting outliers

Now that we have a clustering of the records in the dataset, we can detect the outlying ones. These are the ones who are the farthest from the centroids of their clusters. Track them down and describe any interesting observations that you can make. **(pts 20)** 

```python

def kMeansCentroids (k, data):
    """run kMeans - return centroids"""
    clusters = []
    centroids = []
    
    # initialize clusters
    for i in range(0, k):
        clusters.append([])
        # assign random centroids
        centroids.append(data[random.randint(0, len(data))])
    converged = False
    oldCentroids = None
    while (not converged):
        (clusters, centroids) = untilConvergence(k, data, centroids)
        print 'Current: ',centroids
        print 'Old: ', oldCentroids
        if (oldCentroids):
            converged = testConvergence(centroids, oldCentroids)
        oldCentroids = centroids[:]
    
    return clusters, centroids

def scoreOutlier (point, centroid):
    """Score point in relation to centroid
        low value = close to centroid"""
    xScore = abs(centroid[0]) - abs(point[0])
    yScore = abs(centroid[1]) - abs(point[1])
    return xScore + yScore


clusters, centroids = kMeansCentroids(7, validData)

# find 10 furthest outliers for each cluster
clusterOutliers = []
for i in range(0, len(clusters)):
    centroid = centroids[i]
    cluster = clusters[i]
    outlier = []

    # iterate points of cluster
    for j in range(0, len(cluster)):
        score = scoreOutlier(cluster[j], centroid)
        
        if (len(outlier) < 10):
            # add index of outlier point with it's score
            outlier.append((j, score))
        
        else:
            # test if current point being analyzed is
            # one of 10 furthest outliers from centroid
            for k in range(0, len(outlier)):
                if score > outlier[k][1]:
                    outlier[k] = (j, score)
                    break
                    
    clusterOutliers.append(outlier)


# Print outliers with centroids for comparison    
for i in range(0, len(clusters)):
    cluster = clusters[i]
    centroid = centroids[i]
    outliers = clusterOutliers[i]
    
    print '=' * 15
    print 'Cluster: ', i+1 
    print 'Centroid: ', centroid
    print '-' * 15
    print 'Outliers: '
    for j in range(0, len(outliers)):
        print cluster[outliers[j][0]]
    
    
            
# It is telling to compare the outliers from a variety of clusters.
# Quickly, when comparing the outliers to the cluster's centroid,
# it becomes apparent which clusters are most densely packed.

# For example, cluster 2's outliers barely stray from the 
# centroid value, and are likely a result of a dense cluster
# as opposed to a 'true' outlier data point.
# On the other hand, cluster 4's outliers 2, 3, and 4 are
# all clearly very far away from the centroid. 

# The outlier data by itself can provide insights,
# but paired with the scatterplot it becomes far more useful.
# For instance, without viewing a scatterplot, it may be possible
# for a cluster to have 10 bonafide outliers while still comprising
# 90% of the data sets total points and being extremely dense.
```
